```python
import pandas as pd
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import pickle
```

```python
data = pd.read_csv("spam mail.csv", encoding='latin-1')
```

```python
data = data.dropna(axis=1, how='all')
```

```python
if 'label' not in data.columns:
    data.columns = ['label', 'message']

print("Data loaded successfully!")
print(data.head())
```

```python
def clean_text(text):
    text = text.lower()
    text = ''.join([ch for ch in text if ch not in string.punctuation])
    return text
```

```python
data['message'] = data['message'].astype(str).apply(clean_text)
data['label'] = data['label'].map({'ham': 0, 'spam': 1})
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data['message'], data['label'], test_size=0.2, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
print("\n Model Evaluation:")
print("Accuracy:", round(accuracy_score(y_test, y_pred) * 100, 2), "%")
print(classification_report(y_test, y_pred, target_names=["Ham", "Spam"]))
```

```python
pickle.dump(model, open("spam_model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))
print("\n Model and vectorizer saved successfully!")
```

```python
sample_email = ["Congratulations! You've won a free iPhone. Click here to claim!"]
sample_vec = vectorizer.transform(sample_email)
pred = model.predict(sample_vec)[0]
print("\n Test Email Prediction:")
print("Input:", sample_email[0])
print("Prediction:", "Spam" if pred == 1 else "Not Spam")
```
